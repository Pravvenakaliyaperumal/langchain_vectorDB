````python
import os

# Create directory
os.makedirs("new_articles", exist_ok=True)

files = {
    "google.txt": """Google is investing heavily in generative AI.
    The company announced major updates to its AI research, including new models,
    improved search capabilities, and expanded AI tools for developers.""",

    "ai_startup.txt": """An AI startup recently raised $30 million in its Series B funding round.
    The startup specializes in building scalable AI solutions for enterprises
    and has partnerships with multiple Fortune 500 companies.""",

    "amazon.txt": """Amazon continues to integrate AI across its logistics and retail systems.
    The company is working on drone delivery, automated warehouses,
    and improved recommendation models.""",

    "database.txt": """Modern databases are evolving with vector search capabilities.
    Vector databases like Chroma, Pinecone, and Weaviate enable semantic search
    and power retrieval-augmented generation (RAG) architectures.""",

    "chatgpt.txt": """ChatGPT has transformed natural language processing,
    enabling developers to build chatbots, assistants, and automation systems.
    Recent improvements include better reasoning and tool use.""",

    "microsoft.txt": """Microsoft has invested nearly $10 billion into AI research initiatives.
    The company is expanding Azure AI services and partnering with OpenAI
    to deliver enterprise-grade generative AI solutions."""
}

# Write files
for filename, content in files.items():
    with open(os.path.join("new_articles", filename), "w") as f:
        f.write(content)

print("Dataset created successfully!")
````

````python
loader = DirectoryLoader("new_articles", glob="*.txt", loader_cls=TextLoader)
documents = loader.load()
````

````python
splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)

chunks = splitter.split_documents(documents)
print(chunks)
````

````python
import os
````

````python
!pip install -q sentence_transformers

from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")


vectordb = Chroma.from_documents(
    chunks,
    embedding=embeddings,
    persist_directory="chroma_db"
)

````

````python
vectordb.persist()
````

````python
retriever = vectordb.as_retriever(
    search_type="similarity",
    search_kwargs={"k": 3}
)
````

````python
from langchain_core.prompts import ChatPromptTemplate
````

````python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_template("""
You are a helpful assistant.

Use ONLY the following context to answer the question.
If answer is not in context, say "I don't know."

Context:
{context}

Question:
{question}
""")
````

````python
def rag_answer(question):
    docs = retriever.invoke(question)   # NEW API for LangChain 1.1.2
    context = "\n\n".join([d.page_content for d in docs])

    filled_prompt = prompt.format(context=context, question=question)
    return filled_prompt
````

````python
print(rag_answer("How much money did Microsoft raise?"))
````

````python
from transformers import pipeline

llm = pipeline(
    "text2text-generation",
    model="google/flan-t5-base",
    max_new_tokens=150
)
````

````python
from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_template("""
Use ONLY the following context to answer the question.
If the answer is not in the context, reply: "I don't know."

Context:
{context}

Question: {question}

Answer:
""")
````

````python
history = []

def rag_answer(question, history):
    # 1. Retrieve relevant docs using retriever.invoke()
    docs = retriever.invoke(question)
    context = "\n\n".join([d.page_content for d in docs])

    # 2. Build the final prompt for the LLM
    filled_prompt = prompt.format(
        context=context,
        question=question
    )

    # 3. Generate a final answer using HuggingFace LLM
    answer = llm(filled_prompt)[0]["generated_text"].strip()

    # 4. Save chat history
    history.append({
        "user": question,
        "assistant": answer
    })

    return answer
````

````python
answer = rag_answer("how much AI startup is raised recently?", history)
print("Answer:", answer)
````
